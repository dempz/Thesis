# Using Land/Ocean Data

```python
path = 'C:/Users/Owner/Documents/Thesis/Exploration/Mean Prec Data/Prec_Mean_by_Region.csv'
ocean_land = pd.read_csv(path)

nw = ocean_land['NW_Aus_Prec_Mean'].tolist()
ne = ocean_land['NE_Aus_Prec_Mean'].tolist()
mc = ocean_land['MC_Prec_Mean'].tolist()
io = ocean_land['IO_Prec_Mean'].tolist()
ism = ocean_land['ISM_Prec_Mean'].tolist()
easm = ocean_land['EASM_Prec_Mean'].tolist()
eep = ocean_land['EEP_Prec_Mean'].tolist()
dates = ocean_land['Date'].tolist()
```

```python
def boreal_years(time_series):
    no_nan = [x for x in time_series if math.isnan(x)==False]

    test = []
    for i in range(len(no_nan)):
        year = no_nan[i*12:i*12+12]
        if len(year) == 12:
            test.append(year)
        else:
            break
    
    return test

def austral_years(time_series):
    cropped = time_series[1824+5:] # Need continuous data to take June years consistently. Need this to start at June
    
    test = []
    for i in range(len(cropped)):
        year = cropped[i*12:i*12+12]
        if len(year) == 12:
            test.append(year)
        else:
            break
    
    return test
```

```python
# Measure Austral mean in Borel year and vice versa
def average_jja(time_series):
    jja = [5,6,7] # June, July, August

    years = boreal_years(time_series)
    
    props = []
    for i in range(len(years)):
        observation_year = years[i]
        
        jja_total = 0
        for j in jja:
            jja_total += observation_year[j]
            
        total = sum(observation_year)
        props.append(100*jja_total/total)
    
    return props

# Measure Austral mean in Borel year and vice versa
def average_djf(time_series):
    jja = [6,7,8] # December,January,February

    years = austral_years(time_series)
    
    props = []
    for i in range(len(years)):
        observation_year = years[i]
        
        jja_total = 0
        for j in jja:
            jja_total += observation_year[j]
            
        total = sum(observation_year)
        props.append(100*jja_total/total)
    
    return props
```

```python
easms = average_jja(easm)
isms =average_jja(ism)
nws = average_djf(nw)
nes = average_djf(ne)
mcs = average_djf(mc)
ios = average_djf(io)
eeps = average_djf(eep)

precipitation_props = pd.DataFrame(zip(nws,nes,mcs,ios,eeps,isms,easms),columns=['NW','NE','MC','IO','EEP','ISM','EASM'])
precipitation_props.loc['Mean'] = precipitation_props.mean()
```
